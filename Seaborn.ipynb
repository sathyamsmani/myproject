# Seaborn

## Distribution Plots(Numeric Variables)
- Number columns are called as features
    - distplot - Univiriate analysis(uses 1 feature), 1 number column will be used and the Y axis value will calculated based on the probability(sum of all values equals to 1)
    - joinplot - Biviriate analysis(uses 2 features)
    - pairplot - If we are having more than 2 features tat time we will use this pairplot

## Categorical Plots
- Calculations are based on the attribute values
     - boxplot
     - violinplot
     - countplot - need to give only the x or y axis value (based on the given axis the other axis value will gets calculated)
     - barplot - In this we need to give both the x and y axis values to generate the barplot


```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
sns.jointplot(x='total_bill', y = 'tip', data=df, kind='hex')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.distplot(df['total_bill'])
```

```python
sns.countplot(data=df,y='sex')
```

```python
df['sex'].value_counts()
```

```python
sns.countplot('day',data = df)
```

```python
sns.barplot(data = df, x= 'total_bill', y='sex')
```
